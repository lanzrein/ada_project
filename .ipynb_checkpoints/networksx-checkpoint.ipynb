```python
data = pd.read_csv('data/cleaned_unique_index.csv',low_memory=False)
data.set_index('code',inplace=True)
data.head()
```

```python
origins_france = data['France' == data['countries_en']][['origins_tags']]
origins_france.head()
```

```python
origins_france = origins_france.dropna()
```

```python
z = origins_france["origins_tags"].str.split(',').apply(pd.Series).unstack().dropna().to_frame()
z = z.reset_index().set_index('code')
z.columns = ['','origins_tags']
origins_split = z[['origins_tags']]
```

```python
grouped = origins_split['origins_tags'].value_counts().to_frame()
grouped.columns=['count']
exclude = ['provence','agriculture-ue','ue','lait-origine-france','sud-ouest','hors-france','normandie','hors-union-europeenne','agriculture-ue-non-ue','union-europeenne','bretagne']
for e in exclude : 
    grouped = grouped[grouped.index != e]
top15 = grouped.nlargest(15,'count')
top15
```

```python
import matplotlib as mpl
import math
labs = ['']+list(top15.index)
labs = dict(zip(labs,labs))
g = nx.DiGraph()
i = 1
g.add_star(['']+list(top15.index))

node_sizes = [250*i+250 for i in range(len(g))]
node_sizes
edge_colors=range(2,g.number_of_edges()+2 )
edge_colors =list(map(math.log,top15.values.ravel()))
edge_alphas = [(5+i)/(g.number_of_edges() +4) for i in range(g.number_of_edges())]
pos = nx.layout.spring_layout(g)
nodes = nx.draw_networkx_nodes(g,pos,node_size=300,node_color='#ADD8E6')
labels = nx.draw_networkx_labels(g,pos,labels=labs)
edges = nx.draw_networkx_edges(g,pos,arrowstyle='->',arrowsize=10,width = 2,edge_cmap = plt.cm.winter,edge_color=edge_colors)


pc = mpl.collections.PatchCollection(edges,cmap=plt.cm.winter)
pc.set_array(edge_colors)
plt.colorbar(pc)
ax = plt.gca()
ax.set_axis_off()
fig = plt.gcf()
fig.set_size_inches(12,8)
plt.title("Importation of product to france by count. Log scale is used")
plt.show()
```

```python
z = origins_france.join(data,lsuffix='l',rsuffix='r')
```

```python
z = z[["categories_en"]]
z = z["categories_en"].str.split(',').apply(pd.Series).unstack().dropna().to_frame()
z = z.reset_index().set_index('code')
z.columns = ['','categories_en']
categories_split = z[['categories_en']]
```

```python
by_cat_france = categories_split["categories_en"].value_counts().to_frame()
by_cat_15 = by_cat_france.nlargest(15,'categories_en')
by_cat_15.plot.barh()
```

```python
import matplotlib as mpl
labs = ['']+list(by_cat_15.index)
labs = dict(zip(labs,labs))
g = nx.DiGraph()
i = 1
g.add_star(['']+list(by_cat_15.index))

node_sizes = [100]+list(by_cat_15.values.ravel())
# edge_colors=range(2,g.number_of_edges()+2 )
edge_colors =np.random.randint(10,100,g.number_of_edges())
# edge_alphas = [(5+i)/(g.number_of_edges() +4) for i in range(g.number_of_edges())]
pos = nx.layout.spring_layout(g)
nodes = nx.draw_networkx_nodes(g,pos,node_size=node_sizes,node_color='pink')
labels = nx.draw_networkx_labels(g,pos,labels=labs,font_color='yellow')
edges = nx.draw_networkx_edges(g,pos,node_size=node_sizes,arrowstyle='->',arrowsize=10,width = 2,edge_cmap = plt.cm.winter,edge_color=edge_colors)


pc = mpl.collections.PatchCollection(edges,cmap=plt.cm.winter)
# pc.set_array(edge_colors)
# plt.colorbar(pc)
ax = plt.gca()
ax.set_axis_off()
fig = plt.gcf()
fig.patch.set_facecolor('y')
fig.set_size_inches(14,10)
plt.title("Categories of product to france by count.",color='yellow')
plt.show()
```

```python
## For each main categories do a cutie network graph like above to see if we have any trends :) 
for cat in by_cat_15.index:
    print("------")

    print(cat)
    cat_pd = categories_split[cat ==  categories_split["categories_en"]]
    origin_cat = cat_pd.join(origins_split)
    top10 = origin_cat["origins_tags"].value_counts().to_frame().nlargest(10,"origins_tags")
    do_star_graph_series(top10,mode='edge')
    print("-----------")
```

```python
def do_star_graph_series(series,node_color='pink',font_color='black',mode='node'):
    labs = ['']+list(series.index)
    labs = dict(zip(labs,labs))
    g = nx.DiGraph()
    i = 1
    g.add_star(['']+list(series.index))

    pos = nx.layout.spring_layout(g)

    if mode=='both':
        
        node_sizes = [100]+list(series.values.ravel())
        edge_colors =list(map(math.log,series.values.ravel()))    
        nodes = nx.draw_networkx_nodes(g,pos,node_size=node_sizes,node_color=node_color)
        edges = nx.draw_networkx_edges(g,pos,node_size=node_sizes,arrowstyle='->',arrowsize=10,width = 2,edge_cmap = plt.cm.winter,edge_color=edge_colors)
        pc = mpl.collections.PatchCollection(edges,cmap=plt.cm.winter)
        pc.set_array(edge_colors)
        plt.colorbar(pc)

    # edge_colors=range(2,g.number_of_edges()+2 )
    elif mode=='edge':
        edge_colors =list(map(math.log,series.values.ravel()))
        nodes = nx.draw_networkx_nodes(g,pos,node_color=node_color)
        node_sizes = [100]+list(series.values.ravel())

        edges = nx.draw_networkx_edges(g,pos,node_size=node_sizes,arrowstyle='->',arrowsize=10,width = 2,edge_cmap = plt.cm.winter,edge_color=edge_colors)
        pc = mpl.collections.PatchCollection(edges,cmap=plt.cm.winter)
        pc.set_array(edge_colors)
        plt.colorbar(pc)
    elif mode=='node':
        node_sizes = [100]+list(series.values.ravel())
        nodes = nx.draw_networkx_nodes(g,pos,node_size=node_sizes,node_color=node_color)
        edges = nx.draw_networkx_edges(g,pos,node_size=node_sizes,arrowstyle='->',arrowsize=10,width = 2)

    
    # edge_alphas = [(5+i)/(g.number_of_edges() +4) for i in range(g.number_of_edges())]
#     nodes = nx.draw_networkx_nodes(g,pos,node_size=node_sizes,node_color=node_color)
    labels = nx.draw_networkx_labels(g,pos,labels=labs,font_color=font_color)
#     edges = nx.draw_networkx_edges(g,pos,node_size=node_sizes,arrowstyle='->',arrowsize=10,width = 2,edge_cmap = plt.cm.winter,edge_color=edge_colors)



    ax = plt.gca()
    ax.set_axis_off()
    fig = plt.gcf()
#     fig.patch.set_facecolor('y')
    fig.set_size_inches(14,10)
    plt.title("Categories of product to france by count.")
    plt.show()
```
